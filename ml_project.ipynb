```python
import requests

def download_image_from_drive(url):
   file_id = url.split('id=')[1] 
   download_url = f"https://drive.google.com/uc?export=view&id="
   response = requests.get(download_url+file_id)
   if response.status_code == 200:
     return response.content
   else:
     raise Exception(f"Error downloading image: {response.status_code}")
```

```python
import os

# Initialize empty cache dictionary
image_cache = {}

def get_image_from_drive(url):
    if url in image_cache:
        return image_cache[url]
    else:
        # Download image and store in cache
        image_data = download_image_from_drive(url)
        image_cache[url] = image_data  # Store image data in a list
        return image_data

# ... call get_image_from_drive with your file ID to access and compare the image

# Example usage:
#file_id_1 ="https://drive.google.com/open?id=1uX4yLBpBW1AIUKanafmF9ss1UPuEAM0f"
#file_id_2 = "https://drive.google.com/open?id=1oYDvqqU_oa0NxQEWm-FGTYkAVw8mEK4p"

# Now image_cache will look like this:
# {'your_file_id_1_here': [image_data_1], 'your_file_id_2_here': [image_data_2]}
```

```python
import face_recognition
import cv2

def load_image(file_path):
    image = face_recognition.load_image_file(file_path)
    return image

def get_face_encoding(image):
    face_encoding = face_recognition.face_encodings(image)
    if not face_encoding:
        return None
    return face_encoding[0]

def draw_marker(frame, found):
    frame_height, frame_width, _ = frame.shape
    center_x = int(frame_width / 2)
    center_y = int(frame_height / 2)

    if found:
       # Draw a green tick mark at the center
        tick_size = 80
        thickness = 20
        cv2.line(frame, (center_x - int(tick_size / 2), center_y), (center_x - int(tick_size / 4), center_y + int(tick_size / 2)), (0, 255, 0), thickness)
        cv2.line(frame, (center_x - int(tick_size / 4), center_y + int(tick_size / 2)), (center_x + int(tick_size / 2), center_y - int(tick_size / 2)), (0, 255, 0), thickness)
    else:
        # Draw a red cross at the center
        line_length = 40
        line_thickness = 20
        cv2.line(frame, (center_x - line_length, center_y - line_length), (center_x + line_length, center_y + line_length), (0, 0, 255), line_thickness)
        cv2.line(frame, (center_x - line_length, center_y + line_length), (center_x + line_length, center_y - line_length), (0, 0, 255), line_thickness)




def main():
    # Load the image from the drive
    image_path ="C:\\Users\\rohan\\Downloads\\pexels-lucas-pezeta-2118709 - rohan anvekar.jpg"
    image_from_drive = load_image(image_path)
    face_encoding_from_drive = get_face_encoding(image_from_drive)

    if face_encoding_from_drive is None:
        print("No face found in the provided image.")
        return

    # Open a connection to the camera (you may need to adjust the camera index)
    camera = cv2.VideoCapture(0)

    while True:
        # Capture a frame from the camera
        ret, frame = camera.read()

        # Find face locations and encodings in the current frame
        face_locations = face_recognition.face_locations(frame)
        face_encodings = face_recognition.face_encodings(frame, face_locations)

        # Check if any face is found in the current frame
        if face_encodings:
            # Compare with the face from the image
            match = face_recognition.compare_faces([face_encoding_from_drive], face_encodings[0])

            if match[0]:
                print("Person found!")
            else:
                print("Person on camera does not match the provided image.")

            # Draw the marker on the frame
            draw_marker(frame, match[0])

        # Display the current frame
        cv2.imshow('Camera Feed', frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the camera and close the OpenCV window
    camera.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
